# Contour plot example

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_mark(x, y, color='black'):
    plt.text(x, y, 'X', fontsize=30, color=color)

x = np.arange(1, 6)
y = x
z = [[0, 5, 7, 3, 0],
     [5, 7, 9, 7, 5],
     [0, 7, 9, 10, 4],
     [0, 5, 7, 9, 0],
     [0, 2, 6, 5, 2]]
z.reverse()
marks = [
    (1.2, 4.3),
    (2.0, 4.0),
    (3.2, 3.0),
    (2.8, 3.8),
    (1.8, 2.0),
    (2.1, 1.7),
    (1.5, 1.5),
    (3.0, 3.5),
    (4.5, 1.8),
    (4.5, 1.2),
]

fig, ax = plt.subplots(figsize=(15, 15))
plt.contourf(x, y, z, 10)

for xi, yi in marks:
    plot_mark(xi, yi)
plot_mark(3.8, 2.8, 'red')

ax.set_xticks(x)
ax.set_yticks(y)
plt.show()
```
